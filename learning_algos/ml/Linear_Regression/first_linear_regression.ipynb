```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

# Create the data

### Random data with near a straight line

```python
noise_  = np.random.normal(0, 1, 1000)
```

```python
# y = wTx y = w1x1 + w2x2 + noise

x_range = (-2 , 2)

data_size = 1000

noise_  = np.random.normal(0, 2, 1000)
count = 0

weights_ = [2]
all_values = []
while count<data_size:
    x_ = [np.random.uniform(low=x_range[0], high=x_range[1], size=None)]

    y_ = (weights_ @ np.array(x_) ) + noise_[count]
    x_.append(y_)
    all_values.append( x_)
    count = count+1


```

```python
import matplotlib.pyplot as plt

# Sample data (list of lists)
data = all_values

# Extracting x and y values
x_values, y_values = zip(*data)

# Scatter plot with beautiful styling
plt.figure(figsize=(8, 5), dpi=120)
plt.scatter(x_values, y_values, color='royalblue', edgecolors='black', s=10, alpha=0.75, label="Data Points")

# Beautify plot
plt.xlabel("X-axis", fontsize=12, fontweight='bold')
plt.ylabel("Y-axis", fontsize=12, fontweight='bold')
plt.title("Beautiful Scatter Plot of X vs Y", fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

# Fit the Regression

```python

X = [[x[0]] for x in all_values]
y = [x[1] for x in all_values]


reg = LinearRegression().fit(X, y)
reg.score(X, y)
print(reg.coef_)
print(reg.intercept_)
x_plot = np.linspace(-2, 2, 500).reshape(-1, 1)
y_plot = reg.predict(x_plot)


import matplotlib.pyplot as plt

# Sample data (list of lists)
data = all_values

# Extracting x and y values
x_values, y_values = zip(*data)

# Scatter plot with beautiful styling
plt.figure(figsize=(8, 5), dpi=120)
plt.scatter(x_values, y_values, color='royalblue', edgecolors='black', s=10, alpha=0.75, label="Data Points")
# Plot the regression line
plt.plot(x_plot, y_plot, color='red', linewidth=2, linestyle='-', label="Regression Fit")
# Beautify plot
plt.xlabel("X-axis", fontsize=12, fontweight='bold')
plt.ylabel("Y-axis", fontsize=12, fontweight='bold')
plt.title("Beautiful Scatter Plot of X vs Y", fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
